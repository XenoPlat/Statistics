### Задача 1

#### Даны значения величины заработной платы заемщиков банка $(salary)$ и значения их поведенческого кредитного скоринга $(scoring)$:

$salary = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110]$
$scoring = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]$

#### Без использования сторонних библиотек (вроде numpy, pandas и др.) найдите:
1. ковариацию (смещённую и несмещённую) этих двух величин,
2. коэффициент корреляции Пирсона.

Можно затем посчитать те же значения с использованием библиотек, чтобы проверить себя.

```python
sal = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
sco = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
n = len(sal)

mean_sal = sum(sal) / n
mean_sco = sum(sco) / n

cov_biased = ((sal - mean_sal) * (sco - mean_sco)).sum() / sal.shape[0]
cov_unbiased = ((sal - mean_sal) * (sco - mean_sco)).sum() / (sal.shape[0] - 1)

print('Ковариация смещённая: ', np.round(cov_biased,2))
print('Ковариация несмещённая: ', np.round(cov_unbiased,2))
print('Проверка: ', np.round(np.cov(sal, sco)[0][1],2))
```

```python
std_sal = np.sqrt(sum((sal - mean_sal)**2)/(n-1))
std_sco = np.sqrt(sum((sco - mean_sco)**2)/(n-1))
print('Kоэффициент корреляции Пирсона:',np.round(cov_unbiased/std_sal/std_sco,4))
print('Проверка: ', np.round(np.corrcoef(sal, sco)[0][1],4))
```

### Задача 2

#### Измерены значения IQ выборки студентов, обучающихся в местных технических вузах:
$131, 125, 115, 122, 131, 115, 107, 99, 125, 111$
#### Известно, что в генеральной совокупности IQ распределен нормально. Найдите доверительный интервал для математического ожидания с надежностью 0.95.

```python
iq = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])

iq_mean = np.mean(iq)
iq_std = np.std(iq, ddof=1)

tail = 2.262 * iq_std / len(iq) ** 0.5

print('Доверительный интервал: ',np.round(iq_mean - tail,2), np.round(iq_mean + tail,2))
```

### Задача 3

#### Известно, что рост футболистов в сборной распределён нормально с известной дисперсией $25$. На выборке объёма $27$ выборочное среднее составило $174.2$. Найдите доверительный интервал для математического ожидания с надёжностью $0.95$.

```python
std = np.sqrt(25)
mean = 174.2
n = 27

alpha = np.round((1 - 0.95), 2)
z = -norm.ppf(alpha / 2)

print('Доверительный интервал: ',np.round(mean - z * std / np.sqrt(n), 2), np.round(mean + z * std / np.sqrt(n), 2))
```
