### Задача 1

#### Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

Это задача на сочетания в 2-х независимых множествах: 1) 4 карты; 2) 48 карт.

В 1-м множестве ищется число исходов для комбинаций, когда в нем 1 или 2, или 3, или 4 туза. А во-втором ищется число исходов для комбинаций любых 3-х, 2-х или одного не туза. 

Соответсвующие числа исходов перемножаются, а результаты перемножения складываются между собой, чтобы получить искомый результат. Т.е. число исходов для одного туза из 4-х карт перемножаем с числом исходов для 3-х не тузов из 48 карт и т.д.

Для решения задачи воспользуемся функцией 'combinations' из методички

```python
def combinations(n: int, k: int) -> int:
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))
```

```python
a = combinations(4, 1)*combinations(48, 3)
b = combinations(4, 2)*combinations(48, 2)
c = combinations(4, 3)*combinations(48, 1)
d = combinations(4, 4)*combinations(48, 0)
print(a + b + c + d)
```

### Задача 2

#### Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что рядом будут сидеть: а) два определённых человека? б) три определённых человека?

Исходя из классического определения вероятности, для решения данной задачи необходимо расчитать отношение общего количества (суммы) 'благоприятных' исходов (в нашем случае это когда рядом сидят 2 или 3 определенных человека) к общему количеству исходов (в нашем случае это количество исходов произвольного 'перемешивания' людей на скамейке.

Общее количество исходов является перестановкой и равно n!, т.е. 7!

Для расчета количества 'благоприятных' исходов, необходимо понять что будет элементарным событием. В нашем случае это когда 2 определенных человека сели рядам. При этом следует учесть, что порядок размещения не важен, т.е. комбинация 1-й человек - 2-й человек (1-2) равноценна комбинации 2-1. Аналогично и для 1-2-3 = 1-3-2 = и т.д. Но покамест сосредоточимся на п. 'a)' задачи.

Итак, скамейка на 7 мест. Наша пара может занять 1-е и 2-е место, либо 2-е и 3-е, либо 3-е и 4-е, либо 4-е и 5-е, либо 5-е и 6-е, либо 6-е и 7-е. Итого 6 комбинаций. Т.к. порядок рассадки в паре не вожен, то количество таких комбинаций нужно удвоить, т.е их будет 12.  

Оставшиеся 5 человек могут рассаживаться произвольно и по ним количество исходов будет равно 5! Таким образом, количество благоприятных исходов будет равно 6*2*5!

Для расчета числа перестановок воспользуемся функцией 'permutations' из методички

```python
def permutations(n: int) -> int:
    return np.math.factorial(n)
```

```python
print('Вероятность по п. а) задачи =', 6*2*permutations(5)/permutations(7))
```

Вернемся к условию б) задачи.

Для сокращения объема пояснений, не будем повторять то, что может быть использовано здесь.

Сначала посмотрим на размещение нашей 'троицы' по местам: 1-2-3, 2-3-4, 3-4-5, 4-5-6, 5-6-7. Итого 5 комбинаций. Количество варинатов рассадки, скажем для мест 1-2-3 равно 3! = 6.

Т.о. образом решением для условия б) задачи будет:

```python
print('Вероятность по п. б) задачи =', 5*permutations(3)*permutations(4)/permutations(7))
```

### Задача 4

#### Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?

Событие А = 2,4,6

Событие B = 4,5,6

Изначально вероятности событий A и B одинаковые: P(A) = 3/6, P(B) = 3/6

Обратим внимание, что у A и B есть общие цифры: 4 и 6. Поэтому при наступлении события A с цифрами 4 или 6 вероятность наступления события B будет снижаться и наоборот.

P(A) = 3/6. Допустим  выпала 4. Тогда теперь P(B) становится = 2/6, т.к. может выпасть только 5 или 6.

Следовательно эти 2 события зависимые. Другими словами, существует условная вероятность наступления события B при наступлении события А и она не равна безусловной вероятности события B.

### Задача 5

#### Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.

#### Вопрос: какова вероятность, что вы действительно больны ей?

#### Подсказка: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

Итак дано: А - тест положительный, тогда:

H1: болен

Н2: не болен

Пусть гипотеза 1 (Н1) – болен, гипотеза 2 (Н2) – не болен, тогда:

P1 = 0.001 - болен

Р2 = 0.999 - не болен

Р3 = 0.99 - тест положительный, болезнь есть

Р4 = 0.01 - тест положительный, болезни нет

Тогда полная P(A):

```python
P_A = 0.001 * 0.99 + 0.999 * 0.01
P_A
```

Тогда вероятность, что человек болен при положительном тесте Р(Н1 | тест положит):

```python
0.99 * 0.001 / P_A
```
